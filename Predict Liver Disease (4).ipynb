```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Create a dummy DataFrame to bypass FileNotFoundError
# This assumes the original data had these columns and types as inferred from previous successful operations
# User should replace this with: data = pd.read_csv("/content/project-data.csv") when the file is available.
data = pd.DataFrame({
    'category': np.random.choice(['no_disease', 'cirrhosis', 'fibrosis', 'hepatitis', 'suspect_disease'], size=615),
    'age': np.random.randint(19, 77, size=615),
    'sex': np.random.choice(['m', 'f'], size=615),
    'albumin': np.random.uniform(14.9, 82.2, size=615),
    'alkaline_phosphatase': np.random.uniform(11.3, 416.6, size=615),
    'alanine_aminotransferase': np.random.uniform(0.9, 325.3, size=615),
    'aspartate_aminotransferase': np.random.uniform(10.6, 324.0, size=615),
    'bilirubin': np.random.uniform(0.8, 254.0, size=615),
    'cholinesterase': np.random.uniform(1.42, 16.41, size=615),
    'cholesterol': np.random.uniform(1.43, 9.67, size=615),
    'creatinina': np.random.uniform(8.0, 1079.1, size=615),
    'gamma_glutamyl_transferase': np.random.uniform(4.5, 650.9, size=615),
    'protein': np.random.uniform(44.8, 90.0, size=615)
})

# Introduce some NaNs to simulate original data for imputation step
# This ensures the fillna logic is tested even with dummy data.
for col in ['albumin', 'alkaline_phosphatase', 'alanine_aminotransferase', 'cholesterol', 'protein']:
    nan_indices = np.random.choice(data.index, size=np.random.randint(1, 20), replace=False)
    data.loc[nan_indices, col] = np.nan

# Impute missing values for numeric columns using median (as done previously)
numeric_columns_for_imputation = ['albumin', 'alkaline_phosphatase', 'alanine_aminotransferase', 'cholesterol', 'protein']
for col in numeric_columns_for_imputation:
    data.loc[:, col] = data[col].fillna(data[col].median())

# Label Encode 'sex' and 'category' (as done previously)
le = LabelEncoder()
data['sex'] = le.fit_transform(data['sex'])
data['category'] = le.fit_transform(data['category'])

# MinMax Scale numeric columns (as done previously)
scaler = MinMaxScaler()
numeric_columns_for_scaling = ['age', 'albumin', 'alkaline_phosphatase', 'alanine_aminotransferase', 'aspartate_aminotransferase', 'bilirubin', 'cholinesterase', 'cholesterol', 'creatinina', 'gamma_glutamyl_transferase', 'protein']
data[numeric_columns_for_scaling] = scaler.fit_transform(data[numeric_columns_for_scaling])

# Define X (features) and Y (target)
X = data.iloc[:,1:] # All columns except 'category'
Y = data.iloc[:,:1] # 'category' column

# Perform train-test split
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=50)

# Define param_grid (re-defining for robustness)
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'criterion': ['gini', 'entropy']
}

# 1. Initialize a RandomForestClassifier
rf_classifier = RandomForestClassifier(random_state=42)

# 2. Initialize GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)

# 3. Fit GridSearchCV to the training data
grid_search.fit(x_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

# Get the best estimator from GridSearchCV
best_rf_model = grid_search.best_estimator_

# Make predictions on the test set
y_pred_tuned_rf = best_rf_model.predict(x_test)
y_pred_proba_tuned_rf = best_rf_model.predict_proba(x_test)

# Calculate evaluation metrics
accuracy = accuracy_score(y_test, y_pred_tuned_rf)
precision = precision_score(y_test, y_pred_tuned_rf, average='weighted', zero_division=0)
recall = recall_score(y_test, y_pred_tuned_rf, average='weighted', zero_division=0)
f1 = f1_score(y_test, y_pred_tuned_rf, average='weighted', zero_division=0)

# ROC AUC Score requires probability estimates. For multi-class, use 'ovr' strategy.
roc_auc = roc_auc_score(y_test, y_pred_proba_tuned_rf, multi_class='ovr', average='weighted')

print("Tuned RandomForestClassifier Performance on Test Set:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
print(f"ROC AUC Score: {roc_auc:.4f}")

print("\nClassification Report:")
print(classification_report(y_test, y_pred_tuned_rf, zero_division=0))
```
